```python
#create a table named employee 
import sqlite3 
conn= sqlite3.connect("employees.db")
cursor= conn.cursor()

cursor.execute(""" CREATE TABLE employees(
employee_id INT PRIMARY KEY,
last_name TEXT, 
first_name TEXT,
position TEXT,
salary REAL,
date_hired DATE
""")

print("Table created successfully ")

conn.commit()

```

```python
#write a query to get the unique position from employees table 
sql= """ 
SELECT employees.position 
FROM employees;
"""

cursor.execute(sql)

result= cursor.fetchall()

result

```

```python
import pandas as pd 

pd.DataFrame(result)
```

```python
#write a query to get all employee details from the employee table, order by first name in descending order
sql= """ 
SELECT employees.first_name 
FROM employees
ORDER BY first_name DESC;
"""
cursor.execute(sql)

result= cursor.fetchall()

result
```

```python
import pandas as pd 

pd.DataFrame(result)
```
